```python
#%%
import pandas as pd
import numpy as np

# 1. 模拟一份农产品交易数据（包含各种类型的数据）
data = {
    'product_id': [1001, 1002, 1003, 1004, 1005],                  # 整数
    'product_name': ['Organic Apple', 'Wheat', 'Corn', 'Soybean', 'Avocado'], # 文本
    'price_per_kg': [3.50, 1.25, np.nan, 2.30, 8.99],              # 浮点数 (包含一个空值 NaN)
    'is_export': [True, False, False, True, True],                 # 布尔值
    'harvest_date': ['2026-03-01', '2025-09-15', '2025-10-20', '2025-11-05', '2026-02-10'], # 日期（目前还是文本）
    'grade': ['A', 'C', 'B', 'A', 'A']                             # 评级（适合转为类别型）
}

df = pd.DataFrame(data)

# 2. 核心转换：把文本强行转换为我们需要的特殊数据类型
# 将 harvest_date 从 object 转为真正的日期型 datetime64
df['harvest_date'] = pd.to_datetime(df['harvest_date'])

# 将 grade 从 object 转为类别型 category（按 A, B, C 排序）
df['grade'] = pd.Categorical(df['grade'], categories=['C', 'B', 'A'], ordered=True)

# 3. 将这份完美的数据导出为 CSV 文件到你的桌面 (你可以根据你的电脑路径修改)
# 注意：导出为 CSV 后，特殊类型（如 datetime, category）会被打回原形变成文本，重新导入时需要再次转换。
export_path = '/Users/wangshj/Desktop/agri_data_test.csv' 
df.to_csv(export_path, index=False)
print(f"测试文件已成功保存到: {export_path}\n")

# 4. 见证奇迹的时刻：查看数据类型
print("=== 运行 df.info() 的结果 ===")
df.info()
```
