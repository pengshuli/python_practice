```python
import pandas as pd

# 1. 建立你的当前库存 (Series)
stock = pd.Series([20, 50, 30], index=['D3', 'Magnesium', 'CoQ10'])

# 2. 建立你每天的消耗量 (也是一个 Series)
# 假设你每天吃 1 粒 D3，2 粒镁，1 粒辅酶Q10
daily_dose = pd.Series([1, 2, 1], index=['Magnesium', 'D3','CoQ10'])


print(stock)
print(daily_dose)
```

```python
import pandas as pd
# 每一行都是一个独立的对象
data = [
    {"名称": "D3", "库存": 10, "用量": 1},
    {"名称": "鱼油", "库存": 2, "用量": 2},
    {"名称": "二甲双胍", "库存": 2, "用量": 1, "医生": "张医生"}
]

df = pd.DataFrame(data)
print (df)
```
