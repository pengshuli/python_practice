```python
# 定义查询文件类型及个数的函数，变量为想查询的文件夹路径
def check_files_type(target_folder):
    #创建所有文件的后缀名的字典
    extensions = {}
    #分拆文件名，提取后缀
    for f in os.listdir(target_folder):
        if os.path.isfile(os.path.join(target_folder, f)):
            ext = os.path.splitext(f)[1].lower()
            #将后缀装入字典
            extensions[ext] = extensions.get(ext, 0) + 1
    #打印出后缀名
    print("文件类型统计如下：")
    for ext, count in extensions.items():
        print(f"后缀名: {ext if ext else '无后缀'} | 数量: {count}")     

target_folder = input ("请输入文件夹路径: ")
check_files_type(target_folder)
```

```python
# 检查该文件夹是否存在，若不存在就新建文件夹的函数
def create_new_folders(want_to_creat_path):
    # 如果不存在就新建
    os.makedirs(want_to_creat_path, exist_ok=True)
```

```python
#转移照片、普通视频、低分辨率视频
import os
import shutil

source_dir = r'/Volumes/BAMBI L/哈尔滨 insta360/Camera01'

# 1. 定义分类规则
# 照片后缀
photo_exts = ['.insp', '.jpg', '.jpeg']
# 普通视频后缀 (非全景)
normal_video_exts = ['.mp4', '.mov']
# irv低分辨率视频后缀
low_resolution_exts = ['.lrv']
# bin文件后缀
bin_exts = ['.bin']

# 2. 创建子文件夹路径
photo_dir = r'/Volumes/BAMBI L/哈尔滨 insta360/Photos'
video_dir = r'/Volumes/BAMBI L/哈尔滨 insta360/Normal_Videos'
low_dir = r'/Volumes/BAMBI L/哈尔滨 insta360/Low_Resolution_Videos'
bin_dir = r'/Volumes/BAMBI L/哈尔滨 insta360/Bin'

# 3. 如果文件夹不存在，就建一个
create_new_folders (photo_dir)
create_new_folders (video_dir)
create_new_folders (low_dir)
create_new_folders (bin_dir)

# 4. 开始分家
for f in os.listdir(source_dir):
    file_path = os.path.join(source_dir, f)
    
    # 确保只处理文件，跳过我们刚建好的文件夹,并且提出出他的extension小写名称
    if os.path.isfile(file_path):
        ext = os.path.splitext(f)[1].lower()
        
        # 判断是照片吗？
        if ext in photo_exts:
            shutil.move(file_path, os.path.join(photo_dir, f))
            print(f"已移动照片: {f}")
        
        elif ext in low_resolution_exts:
            shutil.move(file_path, os.path.join(low_dir, f))
            print(f"已移动低分辩率视频: {f}")

        elif ext in bin_exts:
            shutil.move(file_path, os.path.join(bin_dir, f))
            print(f"已移动低分辩率视频: {f}")
            
        # 判断是普通视频吗？
        elif ext in normal_video_exts:
            shutil.move(file_path, os.path.join(video_dir, f))
            print(f"已移动普通视频: {f}")

print("\n--- 预处理完成！现在 Camera01 已归类照片、普通视频、低分辨率视频 ---")
```

```python
import os

source_dir = r'/Volumes/BAMBI L/哈尔滨 insta360/Camera01'

# 1. 准备一个集合来存日期（Set会自动过滤重复，同样的日期只会出现一次）
date_set = set()

# 2. 遍历现在剩下的文件
for f in os.listdir(source_dir):
        parts = f.split('_')
        # 确保切开后的段数足够（防止碰到奇怪命名的文件报错）
        if len(parts) > 1:
            date_str = parts[1] # 拿到 20260405 这一段
            date_set.add(date_str)

# 3. 打印结果
print(f"--- 剩余文件的日期盘点 ---")
print(f"总共有 {len(date_set)} 个不同的拍摄日期：")
# 排序后打印，看起来更清晰
for d in sorted(list(date_set)):
    print(f"日期: {d}")
```

```python
# 独立命名开头带.的报错文件
import os
import shutil

source_dir = r'/Volumes/BAMBI L/哈尔滨 insta360/Camera01'
# 创建一个看得见的文件夹
hidden_box = r'/Volumes/BAMBI L/哈尔滨 insta360/Hidden_Files'

create_new_folders(hidden_box)

# 开始搜寻隐形人
for f in os.listdir(source_dir):
    # 只要是以 . 开头的文件
    if f.startswith('.'):
        old_path = os.path.join(source_dir, f)
        
        # 为了让你能看见，搬过去的时候我们顺便把开头的点给去掉
        # 比如 .VID_... 变成 VISIBLE_VID_...
        new_name = "VISIBLE_" + f.lstrip('.') 
        new_path = os.path.join(hidden_box, new_name)
        
        shutil.move(old_path, new_path)
        print(f"成功揪出隐形文件: {f} -> 已更名为: {new_name}")

print("\n--- 搜寻结束！现在去 Hidden_Files 文件夹看看吧 ---")
```

```python
import os
import shutil

source_dir = r'/Volumes/BAMBI L/哈尔滨 insta360/Camera01'
pano_album_dir = r'/Volumes/BAMBI L/哈尔滨 insta360/Pano_Album'

# 只需要跑这一个循环
for f in os.listdir(source_dir):
    
    file_path = os.path.join(source_dir, f)
    
    # 第一步：切出日期 (只切这一次)
    parts = f.split('_')
    if len(parts) > 1:
        date_str = parts[1]
            
        # 第二步：确定子文件夹路径
        target_sub_dir = os.path.join(pano_album_dir, date_str)
            
        # 第三步：建文件夹 (exist_ok 会自动判断，有就不建，没有就建)
        os.makedirs(target_sub_dir, exist_ok=True)
            
        # 第四步：搬家
        shutil.move(file_path, os.path.join(target_sub_dir, f))
        print(f"已处理: {f} -> {date_str}")
```
