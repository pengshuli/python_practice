```python
#补剂管家“

# 定义supplement这个类，里面需要带有日常吃药这个函数，支持查询补剂的剩余量，以及查询剩余不足三天的量
class Supplement:
    def __init__(self,name,stock,daily_take_in):
        self.name = name
        self.stock = stock 
        self.daily_take_in = daily_take_in

    #日常吃药函数
    def take_in(self):
        if self.stock >= self.daily_take_in:
            self.stock -= self.daily_take_in
            print (f"已记录，今日服用{self.name},{self.daily_take_in}颗。")
            return self.stock
        else:
            print (f"{self.name}仅剩余{self.stock}颗,请及时补充!")
    
    # 查询补剂剩余量
    def inventory_check(self):
        days_left = int(self.stock/self.daily_take_in)
        print(f"{self.name}还可以吃{days_left}天")
        return days_left
    
    # 查询哪些补剂不足三天的量
    def which_inventory (self):
        if self.stock <= 3*self.daily_take_in:
            print (f"{self.name}仅剩余{self.stock}颗, 不足三天的量，请及时补充!")
    
# 定义prescription这个子类，除了上述功能，还需要带有处方药的智能预警
class Prescription(Supplement):
    def __init__(self, name, stock,daily_take_in,doctor_name):
        super().__init__(name,stock,daily_take_in)
        self.doctor_name = doctor_name
    
    #库存告急预警
    def which_inventory (self):
        days_left = int(self.stock/self.daily_take_in)
        if days_left <3:
            print(f"{self.name}仅剩余{self.stock}颗，请及时找开方医生: {self.doctor_name}开药！")
     
# --- 第一阶段：录入药箱 ---
pill_box = {
    "D3": Supplement("D3", 10, 1),
    "鱼油": Supplement("鱼油", 2, 2),
    "二甲双胍": Prescription("二甲双胍", 2, 1, "张医生")
}

today_added = []  # 建立一个空列表，记录今天新增的药

print("=== 补剂管家已启动 ===")
print(f"初始药箱包含: {', '.join(pill_box.keys())}")

while True:
    mode = input("\n请选择操作 (1:添加补剂, 2:添加处方药, start:开始日常使用): ").strip().lower()
    
    if mode == 'start':
        break
        
    name = input("药品名称: ").strip()

    #询问库存
    while True:
        try:
            stock = int(input("当前库存: "))
            break
        except ValueError:
        # 用户未输入数字执行补救措施
            print("❌ 输入无效！请输入数字，不要输入文字或特殊符号。")
    
    #询问每日服用量
    while True:
        try:
            daily = int(input("每日服用量: "))
            break
        except ValueError:
        # 用户未输入数字执行补救措施
            print("❌ 输入无效！请输入数字，不要输入文字或特殊符号。")
    
    if mode == '1':
        pill_box[name] = Supplement(name, stock, daily)
        today_added.append(name)  # 关键点：记下今天添加的药品名字
    elif mode == '2':
        doctor = input("医生姓名: ").strip()
        pill_box[name] = Prescription(name, stock, daily, doctor)
        today_added.append(name)  # 关键点：记下今天添加的药品名字
    else:
        print("无效输入")

# --- 第二阶段：衔接日常交互平台 ---
print("\n" + "="*30)
print("系统就绪！指令说明：")
print("- 输入 [药品名称] 记录服用")
print("- 输入 [check] 查询库存告急药品")
print("- 输入 [q] 退出程序")

while True:
    print("\n" + "="*30)
    print(f"当前药箱已有: {', '.join(pill_box.keys())}")
    
    user_input = input("请选择药品或输入指令: ").strip()

    # 1. 退出+显示出药箱中有哪些药品，以及哪些药品是今日添加的
    if user_input.lower() == 'q':
        print (f"您目前药箱中有以下药品：")
        print(",".join(pill_box.keys()))

        if not today_added:
            print("（今日无新添加）")
        else:
            print (f"您今天添加的药品有：")
            for today in today_added:
                print(f"{today}")
        
        print("程序已退出，祝您健康！")
        break
    
    # 2. 新增功能：查询库存不足三天的药品
    elif user_input.lower() == 'check':
        which_check = input("1.查询库存不足三天药品,2.查询所有药品库存")
        if which_check == '1':
            print("--- 📋 库存告急清单 -----")
            found_low_stock = False
            for pill in pill_box.values():
                # 计算剩余天数
                    pill.which_inventory () # 调用你定义的检查函数
                    found_low_stock = True
        
            if not found_low_stock:
                print("太棒了！目前所有药品库存充足。")
        if which_check == '2':
            print("--- 📋 库存清单 -----")
            for pill in pill_box.values():
                pill.inventory_check()


    # 3. 核心吃药逻辑
    elif user_input in pill_box:
        target_pill = pill_box[user_input]
        target_pill.take_in()
        target_pill.inventory_check()
        
    else:
        print(f"❌ 未识别指令或药品 '{user_input}'。")
```
