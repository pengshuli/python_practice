```python
class Supplement:
    # 1. 构造函数：初始化这个物种的“长相”
    def __init__(self, name, stock):
        self.name = name    # 自己的名字
        self.stock = stock  # 自己的库存

    # 2. 方法：这个物种能做的“动作”
    def take(self):
        if self.stock > 0:
            self.stock -= 1
            print(f"吃了一粒{self.name}，还剩{self.stock}粒。")
        else:
            print(f"{self.name}没货了！")

# 3. 根据蓝图，创造具体的“实例”（Instance）
my_d3 = Supplement("D3", 10)
my_mag = Supplement("Magnesium", 5)

# 4. 让实例干活
my_d3.take()
```
